```python
from sklearn.preprocessing import LabelEncoder

label = LabelEncoder()
for i in ['Sex', 'Hypertension', 'Diabetes', 'Level', 'Fitness Goal', 'Fitness Type']:
  data[i] = label.fit_transform(data[i])
```

```python
from sklearn.preprocessing import StandardScaler

scale = StandardScaler()
data[['Age', 'Height', 'Weight', 'BMI']] = scale.fit_transform(data[['Age', 'Height', 'Weight', 'BMI']])
```

```python
data.head(3)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import random

def get_recommendation():
  print("Kindly enter your details for a personalized workout and diet recommendation")

  user_input = {
      "Sex" : int(input("Enter Sex (Male : 1, Female : 0) : ")),
      "Age" : int(input("Enter Age : ")),
      "Height" : float(input("Enter Height in meter (e.g. 1.75) : ")),
      "Weight" : int(input("Enter Weight in kg : ")),
      "Hypertension" : int(input("Enter Hypertension (Yes : 1, No : 0) : ")),
      "Diabetes" : int(input("Enter Diabetes (Yes : 1, No : 0) : ")),
      "BMI" : float(input("Enter BMI : ")),
      "Level" : int(input("Enter Level (Normal : 0, Obese : 1, Overweight : 2, Underweight : 4) : ")),
      "Fitness Goal" : int(input("Enter Fitness Goal (Weight Gain : 0, Weight Loss : 1) : ")),
      "Fitness Type" : int(input("Enter Fitness Type (Cardio Fitness : 0, Muscular Fitness : 1) : "))
  }

  num_features = ['Age', 'Height', 'Weight', 'BMI']
  user_df = pd.DataFrame([user_input])
  user_df[num_features] = scale.transform(user_df[num_features])

  user_features = data[['Sex', 'Age', 'Height', 'Weight', 'Hypertension', 'Diabetes', 'BMI', 'Level', 'Fitness Goal', 'Fitness Type']]
  similarity_scores = cosine_similarity(user_features, user_df).flatten()

  # Retrieve top similar users and get the first recommendation
  similar_user_indices = similarity_scores.argsort()[-5:][::-1]
  similar_user = data.iloc[similar_user_indices]
  recommendation_1 = similar_user[['Exercises',	'Equipment',	'Diet']].mode().iloc[0]

  # Simulate two additional recommendations by modifying input values slightly
  simulated_recommendations = []
  for _ in range(2):
    modified_input = user_input.copy()
    # Randomly adjust Age, Weight, and BMI with larger variations
    modified_input['Age'] += random.randint(-5, 5)
    modified_input['Weight'] += random.randint(-10, 10)
    modified_input['BMI'] += random.uniform(-0.1, 0.1)
    # Normalize modified input values
    modified_user_df = pd.DataFrame([modified_input]) # Create DataFrame directly
    modified_user_df[num_features] = scale.transform(modified_user_df[num_features])
    # Calculate similarity scores for modified input
    modified_similarity_scores = cosine_similarity(user_features, modified_user_df).flatten()
    modified_similar_user_indices = modified_similarity_scores.argsort()[-5:][::-1]
    modified_similar_users = data.iloc[modified_similar_user_indices]
    recommendation_simulated = modified_similar_users[['Exercises', 'Diet', 'Equipment']].mode().iloc[0]  # Get most common recommendation
    # Check if the recommendation is already in simulated recommendations
    if not any(rec['Exercises'] == recommendation_simulated['Exercises'] and rec['Diet'] == recommendation_simulated['Diet'] and rec['Equipment'] == recommendation_simulated['Equipment'] for rec in simulated_recommendations):
        simulated_recommendations.append(recommendation_simulated) # Append the new recommendation

  print("\nRecommended Workout and Diet Plans based on your input:")
  print("\nRecommendation 1 (Exact match):")
  print("EXERCISES:", recommendation_1['Exercises'])
  print("EQUIPMENTS:", recommendation_1['Equipment'])
  print("DIET:", recommendation_1['Diet'])

  # Iterate through simulated recommendations starting from index 0 for printing
  for idx, rec in enumerate(simulated_recommendations, start=2):
      print(f"\nRecommendation {idx} (Slight variation):")
      print("EXERCISES:", rec['Exercises'])
      print("EQUIPMENTS:", rec['Equipment'])
      print("DIET:", rec['Diet'])

  feedback_matrix = []
  # Iterate through all recommendations (exact match + simulated) for feedback
  all_recommendations = [recommendation_1] + simulated_recommendations
  for i in range(len(all_recommendations)):
      feedback = int(input(f"Was recommendation {i+1} relevant? (Yes : 1, No : 0) : "))
      feedback_matrix.append(feedback)

  relevant_indices = [i + 1 for i, feedback in enumerate(feedback_matrix) if feedback == 1]
  if relevant_indices:
    mrr = np.mean([1 / rank for rank in relevant_indices])  # Calculate MRR
  else:
    mrr = 0.0  # If no relevant recommendations
  print(f"\nMean Reciprocal Rank (MRR): {mrr:.2f}")
  return all_recommendations
```

```python
recommendations = get_recommendation(top_n=3)
```
